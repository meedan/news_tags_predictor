```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
titles = []
links = []
pubdates = []
categories = []
# Sitemap: https://alias.estadao.com.br/sitemap/alias/auto.xml
# Sitemap: https://brasil.estadao.com.br/sitemap/brasil/auto.xml
# Sitemap: https://ciencia.estadao.com.br/sitemap/ciencia/auto.xml
# Sitemap: https://cultura.estadao.com.br/sitemap/cultura/auto.xml
# Sitemap: https://economia.estadao.com.br/sitemap/economia/auto.xml
# Sitemap: https://educacao.estadao.com.br/sitemap/educacao/auto.xml
# Sitemap: https://esportes.estadao.com.br/sitemap/esportes/auto.xml
# Sitemap: https://internacional.estadao.com.br/sitemap/internacional/auto.xml
# Sitemap: https://loterias.estadao.com.br/sitemap/loterias/Sitemap.xml
# Sitemap: https://opiniao.estadao.com.br/sitemap/opiniao/auto.xml
# Sitemap: https://pme.estadao.com.br/sitemap/pme/auto.xml
# Sitemap: https://politica.estadao.com.br/sitemap/politica/auto.xml
# Sitemap: https://sao-paulo.estadao.com.br/sitemap/sao-paulo/auto.xml
# Sitemap: https://saude.estadao.com.br/sitemap/saude/auto.xml
# Sitemap: https://sustentabilidade.estadao.com.br/sitemap/sustentabilidade/auto.xml
# Sitemap: https://www.estadao.com.br/sitemap/www.estadao/auto.xml
# Sitemap: https://esportefera.com.br/sitemap/fera/auto.xml
# Sitemap: https://fotos.estadao.com.br/sitemap/fotos/auto.xml
# c_names = ['ciencia','cultura','economia','educacao','esportes','internacional','loterias','opiniao','pme','politica','sao-paulo','saude','sustentabilidade']
c_names = ['ciencia','cultura','economia','educacao','esportes','internacional','pme','politica','saude','sustentabilidade']

for category in c_names:
  for year in range(2022,2023):
    for month in range(1,13):
      try:
        # r_url = "https://"+category+".estadao.com.br/sitemaps/auto/2022/"+f"{month:02d}"+"/sitemap-news.xml"
        # print(r_url)
        r = requests.get("https://"+category+".estadao.com.br/sitemaps/auto/"+str(year)+"/"+f"{month:02d}"+"/sitemap-news.xml")
        # return 
        

        soup = BeautifulSoup(r.content, features='lxml')
        print(soup)
        urls = soup.findAll('url')      

        for a in urls:
          # print(a)
          title = a.find('news:title').text
          link = a.find('loc').text
          published = a.find('lastmod').text
          # category = a.find('news:genres').text
          # print(category)
          titles.append(title)
          links.append(link)
          pubdates.append(published)
          categories.append(category)

          # titles.append(title)
          # links.append(link)
          # pubdates.append(published)
          # cats.append(category)

            # article = {
            #     'title': title,
            #     'link': link,
            #     'pubDate': published
            #     }
            # article_list.append(article)
        # print(article_list)
      except Exception as e:
        print('The scraping job failed. See exception: ')
        print(e)
```

```python
df = pd.DataFrame({'title': titles,'link': links,'pubDate': pubdates,'category':categories})
df
```

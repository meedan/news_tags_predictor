```python
# folder_names = ['businessstandard.xml','economictimes.xml','firstpost.xml','indiatoday.xml','thehindu1.xml','thehindu2.xml']
folder_names = ['businessstandard.xml','economictimes.xml','firstpost.xml','indiatoday.xml']
# folder_names  = ['indiatoday.xml']
filenames = []
for year in range(2021,2023):
    s_year = f'{year:04d}'
    for month in range(1,13):
        s_month = f'{month:02d}'
        for day in range(1,32):
            for x in folder_names:
                s_day = f'{day:02d}'
                year_folder= 'data/'
                year_folder += s_year+"-"+s_month+"-"+s_day+"/"
                year_folder+=x
                print(year_folder)
                filenames.append([year_folder,x])
# for i in range(3,30):
# #     print("2022-03-%02d/" % (i,))
    
#     for x in folder_names:
#         year_folder = "2022-03-%02d/" % (i,)

#         year_folder+=x
# #         print(filenames)
#         filenames.append([year_folder,x])
```

```python
import os.path

titles = []
locs = []
publication_dates = []
keywords = []
source_files = []

for filename in filenames:
    if os.path.exists(filename[0]):
        tree = ET.parse(filename[0])
        root = tree.getroot()

        for url in root:
            source_files.append(filename[1])

            for a_url in url:

                if a_url.tag.find('loc')>=0:
                        locs.append(a_url.text)
                for news in a_url:
        #             print(news.tag, news.attrib)
                    if news.tag.find('title')>=0:
                        titles.append(news.text)
                    if news.tag.find('publication_date')>=0:
                        publication_dates.append(news.text)
                    if news.tag.find('keywords')>=0:
                        keywords.append(news.text)
```

```python
import pandas as pd
# df = pd.DataFrame({"title":titles,"loc":locs,"publication_date":publication_dates})#,"keyword":keywords})
df = pd.DataFrame({"title":titles,"loc":locs,"publication_date":publication_dates,"source_file":source_files})#,"keyword":keywords})

df
```
